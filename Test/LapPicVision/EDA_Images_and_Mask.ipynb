```python
cv2.imshow('Mask', ImgMask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
nonzero = np.count_nonzero(ImgMask)
zero = ImgMask[np.where(ImgMask == 0)].size
total_params = ImgMask.shape[0]*ImgMask.shape[1]*ImgMask.shape[2]

print("Nonzero count: ", nonzero)
print("Zero count: ", zero)

print("Dimensions: ", ImgMask.shape)
print("Total number of pixels: ", total_params)
assert total_params == zero + nonzero
```
